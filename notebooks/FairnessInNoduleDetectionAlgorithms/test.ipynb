```python
import bibtexparser
import csv

# Function to convert .bib file to .csv
def bib_to_csv(bib_file, csv_file):
    # Read the .bib file
    with open(bib_file, 'r') as bibtex_file:
        bib_database = bibtexparser.load(bibtex_file)
    
    # Get the entries
    entries = bib_database.entries
    
    # Get the fieldnames from the first entry
    fieldnames = set()
    for entry in entries:
        fieldnames.update(entry.keys())
    fieldnames = sorted(fieldnames)  # Sort the fieldnames for consistency

    # Write to the .csv file
    with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for entry in entries:
            writer.writerow(entry)

# Example usage
bib_file = '/Users/john/Projects/SOTAEvaluationNoduleDetection/Papers/ConnectedPapers-for-SANet%3A-A-Slice%20Aware-Network-for-Pulmonary-Nodule-Detection.bib'  # Replace with your .bib file path
csv_file = '/Users/john/Projects/SOTAEvaluationNoduleDetection/Papers/ConnectedPapers-for-SANet%3A-A-Slice%20Aware-Network-for-Pulmonary-Nodule-Detection.csv'   # Replace with your desired .csv file path
bib_to_csv(bib_file, csv_file)
```
